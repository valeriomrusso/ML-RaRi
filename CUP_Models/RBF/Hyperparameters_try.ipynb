```python
import keras_tuner as kt
from tensorflow import keras

from load_cup import load_cup_data
from build_rbf_fixed import build_rbf_model
from plotting_cup import plot_cup
```

```python
X_train, y_train, X_val, y_val, X_test, y_test = load_cup_data()

# Parametri fissi
input_dim = 12
output_dim = 3

n_centers = 120 #Lower value more flat
gamma = 0.17 #Higher value more flat - Lower value lower val_loss
batch_size = 20 #Higher more flat - low value low test_loss
learning_rate = 0.004 #Lower value more flat - high value flatline - high value low test_loss high val_loss

# Creazione del modello
rbf_model = build_rbf_model(input_dim, output_dim, n_centers, gamma, learning_rate)

# Addestramento del modello
history = rbf_model.fit(
    X_train, y_train,
    epochs=200,
    batch_size=batch_size,
    validation_data=(X_val, y_val),
    callbacks=[keras.callbacks.EarlyStopping('val_loss', patience=5)]
)

plot_cup(history)
```

```python
import matplotlib.pyplot as plt
import json
import os
from itertools import islice

path = "results/results007"

# Percorso al file results.json
results_file = path + "/results.json"

# Carica i dati da results.json
with open(results_file, "r") as f:
    results_data = json.load(f)

# Ottieni i dati necessari
hyperparameters = results_data["Hyperparameters"]
test_loss = results_data["Test Loss"]
test_mse = results_data["Test MSE"]

# Testo da visualizzare
text_info = (
    f"Hyperparameters:\n"
    + "\n".join([f"  {key}: {value}" for key, value in dict(islice(hyperparameters.items(), 4)).items()])
)
test_results = f"Test Results:\n  Test Loss: {test_loss:.20f}\n  Test MSE: {test_mse:.20f}"

# Percorsi dei grafici di training
training_loss_plot_path = path + "/training_loss_plot.png"
training_mse_plot_path = path + "/training_mse_plot.png"

# Carica le immagini dei grafici
loss_img = plt.imread(training_loss_plot_path)
mse_img = plt.imread(training_mse_plot_path)

# Crea la figura
fig, ax = plt.subplots(2, 2, figsize=(12, 8), gridspec_kw={'height_ratios': [1, 2]})

# Disattiva gli assi per la prima riga e inserisci il testo
ax[0, 0].axis("off")
ax[0, 1].axis("off")

# Inserisci i titoli pi grandi con dimensione maggiore per i titoli
ax[0, 0].text(0, 0.9, "Hyperparameters:", fontsize=16, va="top", ha="left", fontweight="bold")
ax[0, 0].text(0, 0.7, text_info, fontsize=12, va="top", ha="left", wrap=True)

ax[0, 1].text(0, 0.9, "Test Results:", fontsize=16, va="top", ha="left", fontweight="bold")
ax[0, 1].text(0, 0.7, test_results, fontsize=12, va="top", ha="left", wrap=True)

ax[0, 0].set_xlim(0, 1)
ax[0, 0].set_ylim(0, 1)
ax[0, 1].set_xlim(0, 1)
ax[0, 1].set_ylim(0, 1)

# Mostra i grafici nella seconda riga
ax[1, 0].imshow(loss_img)
ax[1, 0].axis("off")
ax[1, 0].set_title("Training Loss Plot")

ax[1, 1].imshow(mse_img)
ax[1, 1].axis("off")
ax[1, 1].set_title("Training MSE Plot")

# Regola il layout
plt.tight_layout()

# Salva l'immagine combinata
combined_image_path = "results/combined_results007.png"
plt.savefig(combined_image_path, dpi=300)
plt.show()

print(f"Immagine combinata salvata in: {combined_image_path}")
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Elenco dei percorsi delle immagini
image_paths = [f"results/combined_results{str(i).zfill(3)}.png" for i in range(1, 8)]

# Crea una figura con un layout di 2 righe e 4 colonne
fig, axes = plt.subplots(2, 4, figsize=(16, 8))

# Flatten la matrice degli assi per un accesso pi semplice
axes = axes.flatten()

# Cicla sulle immagini e carica ciascuna immagine
for i, image_path in enumerate(image_paths):
    # Carica l'immagine
    img = mpimg.imread(image_path)

    # Mostra l'immagine
    ax = axes[i]
    ax.imshow(img)
    ax.axis("off")  # Disabilita gli assi

    # Aggiungi il testo sopra l'immagine
    ax.text(0.5, 1.05, f"Training n {i + 1}", fontsize=14, ha="center", va="bottom", fontweight="bold", transform=ax.transAxes)

# Rimuovi l'asse vuoto dell'ultima posizione
axes[-1].axis('off')

# Regola il layout per evitare sovrapposizioni
plt.tight_layout()

# Salva l'immagine combinata
combined_image_path = "results/combined_results_all_trainings.png"
plt.savefig(combined_image_path, dpi=300)
plt.show()

print(f"Immagine combinata di tutti i training salvata in: {combined_image_path}")
```
